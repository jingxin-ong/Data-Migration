```python
import pandas as pd
import sys
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://postgres:password@localhost:5432/mydatabase')
```

### Converting CSV into SQL table  

```python
df = pd.read_csv('cleaned_fp_oct2020(with_kcc_answer).csv')
df.to_sql('farmpulse', engine)
```
